```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import ast
```

```python
zno_results = pd.read_csv('results/zno-results.csv',delimiter=',')
folktables_results = pd.read_csv('results/folktables-results.csv',delimiter=',')
students_loans_diabetes_results = pd.read_csv('results/students-loans-diabetes-results.csv',delimiter=',')
```

```python
results_df_raw = pd.concat([zno_results, folktables_results, students_loans_diabetes_results], ignore_index=True, axis=0)
```

```python
#results_df_raw = pd.read_csv('results/zno-sensitivty-analysis-results.csv',delimiter=',')
```

```python
# Make it pretty

results_df = results_df_raw.copy()

def prettyify_dict(x):
    temp_dict = dict(sorted(ast.literal_eval(x).items()))
    temp_str = str(temp_dict)
    temp_str = temp_str.replace('{','')
    temp_str = temp_str.replace('}','')
    temp_str = temp_str.replace("'",'')
    temp_str = temp_str.replace(',',';')
    return temp_str

dataset_mapper = {'Ukr': 'EIE (Ukranian)',
                  'Hist': 'EIE (History)',
                  'Math': 'EIE (Math)',
                  'Phys': 'EIE (Physics)',
                  'Chem': 'EIE (Chemsitry)',
                  'Geo': 'EIE (Geography)',
                  'Eng': 'EIE (English)',
                  'Deu': 'EIE (German)',
                  'ACSEmployment': 'Folktables (Employment)',
                  'ACSIncome': 'Folktables (Income)',
                  'ACSPublicCoverage': 'Folktables (Public Medical Coverage)',
                  'ACSMobility': 'Folktables (Mobility)',
                  'ACSTravelTime': 'Folktables (Travel Time)',
                  'loans': 'Loan Assessment',
                  'diabetes': 'Diabetes Risk Assessment',
                  'students': 'Student Performance'}

sensitve_attr_mapper = {
                        'education': 'Education Level',
                        'parents_education': "Parent's Education Level"}


results_df['dataset'] = results_df['dataset'].map(dataset_mapper)
results_df['sensitive_attr'] = results_df['sensitive_attr'].replace(sensitve_attr_mapper)

results_df['sensitive_attr'] = results_df['sensitive_attr'].apply(lambda x: str(x).capitalize())

# After renames
results_df['group_dist'] = results_df['group_dist'].apply(prettyify_dict)
results_df['group_prev'] = results_df['group_prev'].apply(prettyify_dict)

# Lastly rename columns:
column_mapper = {'dataset': 'Dataset (Outcome)',
                 'sample_size': 'Number of Samples',
                 'overall_prev': 'Overall Prevelance (%)',
                 'sensitive_attr': 'Sensitive Attribute',
                 'group_dist': 'Group Distribution (%)',
                 'group_prev': 'Group Prevelance (%)',
                 'max_group_prev_diff': 'Maximum Group Prevelance Difference (%)'}

results_df.rename(columns=column_mapper, inplace=True)
```
